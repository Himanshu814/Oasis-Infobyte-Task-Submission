Visualisation

```python
#HeatMap
plt.style.use('seaborn-ticks')
plt.figure(figsize=(12, 10))
sns.heatmap(dataset.corr(), annot=True,cmap='autumn')
plt.show()
```

```python
sns.pairplot(dataset)
```

```python
#unemployment rate according to different regions of India
dataset.columns= ["State","Date","Frequency","EUR","EE","ELPR","Region","longitude","latitude"]

plt.figure(figsize=(10, 8))
plt.title("Unemployment Rate")
sns.histplot(x="EUR", hue="Region", data=dataset)
plt.show()
```
